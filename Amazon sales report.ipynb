# Group by function in pandas is used to roup data based on one or more columns in a dataframe

```python
df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
QTY = df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
sns.barplot(x='Size', y='Quantity', data = QTY)
```

```python
# histogram
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str) # category data type change

column_data = df['Category'] 
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=30, edgecolor='black') # bins 30 is same distance
plt.show()
```

```python
B2B_check = df['B2B'].value_counts()

# Plot the pie chart
plt.pie(
    B2B_check, 
    labels=B2B_check.index,  # Use the category names as labels
    autopct='%1.1f%%',       # Display percentages
    startangle=90            # Rotate the chart for better visualization
)
plt.title("B2B Data Distribution")  # Add a title for clarity
plt.show()
```

```python
# prepare data for scatter plot
xdata = df['Category']
ydata = df['Size']

#PLOT THE SCatter plot
plt.scatter(xdata, ydata)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('scatter plot')
plt.show()
```

```python
#plot count of cities by state

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
# top 10 states

top10state = df['ship-state'].value_counts().head(10)
#plot count of cities by store
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top10state.index)],x='ship-state')
plt.ylabel('count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```
